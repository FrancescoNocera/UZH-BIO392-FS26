## Introduction to pandas and NumPy

Before working with real biological data, it is useful to learn two very common Python libraries:

- **pandas**: for working with tabular data (https://pandas.pydata.org/docs/)
- **NumPy**: for working with numbers and arrays (https://numpy.org/doc/)

These libraries are widely used in data analysis and will be helpful for the next notebook on Progenetix data.

```python
# Import Pandas and Numpy
import pandas as pd
import numpy as np
```

A pandas DataFrame is a table with rows and columns.
It is one of the most common data structures in Python data analysis.

```python
# Create a simple dataset and use pandas dataframe for it
data = {
    "sample_id": ["S1", "S2", "S3"],
    "cancer_type": ["breast", "lung", "colon"],
    "value": [2.3, 1.8, 3.1]
}

df = pd.DataFrame(data)
df
```

```python
df.columns
```

```python
df.info()
```

```python
# select one column
df["cancer_type"]
```

```python
# select more than one column
df[["sample_id", "value"]]
```

This command exports the DataFrame to a CSV file.
The option `index=False` means that the row index is not written to the file.

```python
df.to_csv("./output/20260415_dataset.csv", index=False)
```

Read the created dataset
