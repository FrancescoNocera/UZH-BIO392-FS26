```python
!bwa mem -M -t 2 \
    -R "@RG\tID:A\tPL:ILLUMINA\tLB:lib1\tSM:A" \
    ./reference/chr14.fa \
    ./output/fastp/clean_A_1.fq.gz \
    ./output/fastp/clean_A_2.fq.gz | samtools sort -@ 2 -o ./output/bwa/A_sorted.bam -
```

## Mark Duplication

Use GATK `MarkDuplicates` command whcih build on top of `picard`   

```python
import os
os.environ["JAVA_HOME"] = "/Users/noel/miniconda3/envs/bio392/lib/jvm"
os.environ["PATH"] = os.path.join(os.environ["JAVA_HOME"], "bin") + ":" + os.environ["PATH"]

!gatk --java-options "-Xmx4G" MarkDuplicates \
        -I ./output/bwa/A_sort.bam \
        -O ./output/bwa/A_sort_markdup.bam \
        -M ./output/bwa/A_sort_markdup_metrics.txt \
        --TMP_DIR .
```

## Base Quality Score Recalibration (BQSR)

Before we jump in to BQSR, we need addition preparation on reference sequence.

Verify that you already create index of reference fasta file with `samtools faidx`

```python
!samtools faidx ./reference/chr14.fa
```

Second, create a sequence dictionary for reference sequence with `gatk CreateSequenceDictionary`

```python
!gatk --java-options "-Xmx4G" CreateSequenceDictionary \
    -R ./reference/chr14.fa
```

Create output folder for BQSR

```python
!mkdir -p ./output/bqsr
```

We ready for Base Quality Score Recalibration (BQSR)

    1. Build model with command `gatk BaseRecalibrator`
    
    2. Apply model to our data with command `gatk ApplyBQSR`

```python
!gatk --java-options "-Xmx4G" BaseRecalibrator \
        -R ./reference/chr14.fa \
        -I ./output/bwa/A_sort_markdup.bam \
        -O ./output/bqsr/A_sort_markdup_BQSR.report \
        --known-sites ./db/Homo_sapiens_assembly38.dbsnp138_roi.vcf.gz \
        --known-sites ./db/resources_broad_hg38_v0_Homo_sapiens_assembly38.known_indels.vcf.gz \
        --known-sites ./db/resources_broad_hg38_v0_Mills_and_1000G_gold_standard.indels.hg38.vcf.gz
```

```python
!gatk --java-options "-Xmx4G" ApplyBQSR \
        -I ./output/bwa/A_sort_markdup.bam \
        -O ./output/bqsr/A_sort_markdup_recal.bam \
        --bqsr-recal-file ./output/bqsr/A_sort_markdup_BQSR.report
```

Finally we got our Ready Analysis BAM for Variant calling.
