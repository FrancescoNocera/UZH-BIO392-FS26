Two structural variants were identified in the same region of chromosome 14 (around 104.9 Mb), both passing filters despite being imprecise calls: a ~5.5 kb duplication (DUP) supported by 17 paired-end reads, and a ~440 bp deletion (DEL) supported by 40 paired-end reads with a higher quality score of 744. The co-occurrence of a DUP and DEL in the same locus is notable — this could reflect a complex rearrangement, or the DUP may be a false positive arising from the abnormal read depth signature created by the nested deletion. Both calls have relatively low mapping quality (MAPQ 22–24) and wide confidence intervals (±50 bp on both breakpoints), so experimental validation would be warranted before drawing strong conclusions.
